```python
import time

# Constants to represent traffic light states
RED = "RED"
GREEN = "GREEN"
YELLOW = "YELLOW"

# TrafficLight class to handle light states
class TrafficLight:
    def __init__(self, location):
        self.location = location  # e.g., "North-South" or "East-West"
        self.state = RED  # Default state is red

    # Function to change the light's state
    def change_state(self, new_state):
        self.state = new_state
        print(f"Traffic Light at {self.location} is now {self.state}")

    # Function to simulate traffic light cycle
    def run(self):
        while True:
            if self.state == RED:
                time.sleep(5)  # Red light for 5 seconds
                self.change_state(GREEN)
            elif self.state == GREEN:
                time.sleep(5)  # Green light for 5 seconds
                self.change_state(YELLOW)
            elif self.state == YELLOW:
                time.sleep(2)  # Yellow light for 2 seconds
                self.change_state(RED)

# PedestrianButton class to simulate pedestrian crossing request
class PedestrianButton:
    def __init__(self):
        self.is_pressed = False

    # Function to simulate pressing the button
    def press(self):
        self.is_pressed = True
        print("Pedestrian button pressed!")

    # Function to reset the button (after lights change)
    def reset(self):
        self.is_pressed = False

# Function to simulate the intersection
def intersection_simulation():
    light1 = TrafficLight("North-South")  # Traffic light for North-South direction
    light2 = TrafficLight("East-West")  # Traffic light for East-West direction
    button = PedestrianButton()  # Pedestrian crossing button

    # Run the traffic lights and simulate button press
    while True:
        # Simulate regular traffic light cycle
        if not button.is_pressed:
            print("\n--- Normal Traffic Flow ---")
            light1.run()
            light2.run()

        # If pedestrian button is pressed, stop traffic flow for pedestrians
        if button.is_pressed:
            print("\n--- Pedestrian Request ---")
            light1.change_state(RED)
            light2.change_state(RED)
            time.sleep(3)  # Give 3 seconds for pedestrians to cross
            button.reset()  # Reset the button after the crossing
            light1.change_state(GREEN)  # Restore green light for traffic
            light2.change_state(GREEN)

if __name__ == "__main__":
    intersection_simulation()
```
